**Rename education status**


```python
def replace_education(x):
    if x == "Haven't Finished High School (less than 13 years ed)":
        return "below high-school"
    elif x == "Didn't Finish High School (less than 13 years ed)":
        return 'below high-school'
    elif x == "High School Degree (12-13 years)":
        return 'high school passed'
    elif x == "Some Undergrad (higher ed)":
        return 'undergrad studying'
    elif x == "Undergraduate Degree (3-5 years higher ed)":
        return 'undergrad passed'
    elif x == "Some Graduate School":
        return 'grad studying'
    elif x == "Graduate Degree":
        return 'grad passed'
    else:
        return 'unknown'
df['education'] = df['education'].apply(replace_education)
```

Convert float values to int values

```python
def func(x):
    if np.isnan(x):
        return x
    else:
        return int(x)
df['house_Eng'] = df['house_Eng'].apply(func)
```

```python
df['house_Eng'] = df['house_Eng'].fillna(-1).astype(int)
```

**6. Identify relationship between number of years spent in English speaking country and language efficiency**

Among the non-native speaker, I want to see whether the number of years spent in English speaking country has significant influence on the language efficiency.

```python
non_natives['Eng_country_yrs'].value_counts()
```

Remove participants with less than 60 years of residence in English speaking country, because frequency of these participants is very low (<150).

```python
eng_country_data = non_natives[non_natives['Eng_country_yrs'] <= 60]
```

**Graph 2**

This graph shows the average correct score depending on the number of years spent in the English speaking coutry.
In this graph, we see a strong correlation between the two variables which shows that the more years an individual spends in English speaking country the more profienct s/he is expected to be.


```python
ax = eng_country_data.groupby('Eng_country_yrs')['correct'].mean().plot( color = 'red')
ax.set(xlabel="Number of years spent in English speaking country", ylabel="Average correct score")
```

**Effect of living with a native English speaker**
